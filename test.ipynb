```python
from aloecrypt._plugin import _wire_pack, _Plugin
from aloecrypt.api import *
from aloecrypt.functions import perform_handshake, _ts_bytes_now
from datetime import datetime

plugin = _Plugin()

print("=" * 60)
print("ALOECRYPT COMPREHENSIVE TEST SUITE")
print("=" * 60)

# ── 1. Identity Creation ─────────────────────────────────────

print("\n── 1. Identity Creation ──")

root_a = DilithiumSigner.new()
root_b = DilithiumSigner.new()
print(f"  root_a address: {root_a.address[:8].hex()}...")
print(f"  root_b address: {root_b.address[:8].hex()}...")
assert root_a.is_root == True
assert root_b.is_root == True
print("  ✓ Root signers created and are_root")

# ── 2. Delegate Creation ─────────────────────────────────────

print("\n── 2. Delegate Creation ──")

EMPTY_TS = b'\x00' * 8
delegate_a = root_a.create_dilithium_signer(EMPTY_TS, EMPTY_TS, 0, 0)
delegate_b = root_b.create_dilithium_signer(EMPTY_TS, EMPTY_TS, 0, 0)
assert delegate_a.is_root == False
assert delegate_b.is_root == False
assert delegate_a.auth_address == root_a.address, "delegate auth_address matches root address"
assert delegate_b.auth_address == root_b.address, "delegate auth_address matches root address"
assert delegate_a.auth_address == root_a.address, "delegate authorized by root"
assert delegate_b.auth_address == root_b.address, "delegate authorized by root"
print(f"  delegate_a address: {delegate_a.address[:8].hex()}... (matches root)")
print(f"  delegate_b address: {delegate_b.address[:8].hex()}... (matches root)")
print("  ✓ Delegates created, non-root, share root address")

# ── 3. KEM Creation ──────────────────────────────────────────

print("\n── 3. KEM Creation ──")

kem_a = delegate_a.create_kyber_kem(EMPTY_TS, EMPTY_TS, 0, 0)
print(f"  kem_a address: {kem_a.address[:8].hex()}...")
canonical_kem = delegate_a.canonical_kyber_kem(b'\x00', EMPTY_TS, EMPTY_TS, 0, 0)
canonical_kem2 = delegate_a.canonical_kyber_kem(b'\x00', EMPTY_TS, EMPTY_TS, 0, 0)
assert canonical_kem.kyb_pubkey == canonical_kem2.kyb_pubkey, "canonical KEMs are deterministic"
print("  ✓ Random and canonical KEMs created")

# ── 4. Signing & Verification ────────────────────────────────

print("\n── 4. Signing & Verification ──")

message = b"test message for signing"
signature = delegate_a.sign(message)
assert len(signature) == 3309, f"signature length {len(signature)}"
print(f"  signature: {signature[:8].hex()}... ({len(signature)} bytes)")

verifier_a = DilithiumVerifier.from_dilithium_signer(delegate_a)
assert verifier_a.address == delegate_a.address
print(f"  verifier address matches: ✓")

# ── 5. Wire Serialization Round-trips ────────────────────────

print("\n── 5. Wire Serialization Round-trips ──")

for name, obj in [
    ("DilithiumSigner", root_a),
    ("DilithiumVerifier", verifier_a),
    ("KyberFullKEM", kem_a),
    ("KyberPublicKEM", kem_a.to_kyber_public_kem()),
    ("SessionBuilder.empty", SessionBuilder.empty()),
]:
    wire = obj.to_wire_bytes()
    restored = type(obj).from_wire_bytes(wire)
    assert wire == restored.to_wire_bytes(), f"{name} wire round-trip failed"
    print(f"  {name}: {len(wire)} bytes ✓")

# ── 6. Empty Round-trips (all types) ─────────────────────────

print("\n── 6. Empty + to_bytes + from_bytes ──")

empty_types = [
    ("KyberFullKEM", KyberFullKEM),
    ("XKyberFullKEM", XKyberFullKEM),
    ("KyberPublicKEM", KyberPublicKEM),
    ("PartyINTRO", PartyINTRO),
    ("PartyCIPHER", PartyCIPHER),
    ("FullCIPHER", FullCIPHER),
    ("CounterPartySECRET", CounterPartySECRET),
    ("PartyCHALLENGE", PartyCHALLENGE),
    ("PartyRESPONSE", PartyRESPONSE),
    ("CounterPartyCHALLENGE", CounterPartyCHALLENGE),
    ("FromSecretsInput", FromSecretsInput),
    ("SessionBuilder", SessionBuilder),
    ("MsgHELLO", MsgHELLO),
    ("MsgSYN", MsgSYN),
    ("MsgACK", MsgACK),
    ("MsgSYNACK", MsgSYNACK),
    ("MsgWELCOME", MsgWELCOME),
    ("MsgGOODBYE", MsgGOODBYE),
    ("MsgRETRY", MsgRETRY),
    ("MsgRESYN", MsgRESYN),
    ("Party", Party),
    ("CounterParty", CounterParty),
    ("AloecryptSession", AloecryptSession),
    ("XParty", XParty),
    ("XCounterParty", XCounterParty),
    ("XAloecryptSession", XAloecryptSession),
    ("XDilithiumSigner", XDilithiumSigner),
    ("DilithiumSigner", DilithiumSigner),
    ("DilithiumVerifier", DilithiumVerifier),
]

for name, cls in empty_types:
    obj = cls.empty()
    assert isinstance(obj, cls), f"{name}.empty() type mismatch"
    tb = obj.to_bytes
    restored = cls.from_bytes(tb)
    assert isinstance(restored, cls), f"{name}.from_bytes() type mismatch"

print(f"  ✓ All {len(empty_types)} types: empty + to_bytes + from_bytes")

# ── 7. PEM Serialization ─────────────────────────────────────

print("\n── 7. PEM Serialization ──")

pem_types = [
    ("XKyberFullKEM", XKyberFullKEM),
    ("KyberPublicKEM", KyberPublicKEM),
    ("XAloecryptSession", XAloecryptSession),
    ("MsgHELLO", MsgHELLO),
    ("MsgSYN", MsgSYN),
    ("MsgACK", MsgACK),
    ("MsgSYNACK", MsgSYNACK),
    ("MsgWELCOME", MsgWELCOME),
    ("MsgGOODBYE", MsgGOODBYE),
    ("MsgRETRY", MsgRETRY),
    ("MsgRESYN", MsgRESYN),
    ("XDilithiumSigner", XDilithiumSigner),
    ("DilithiumVerifier", DilithiumVerifier),
]

for name, cls in pem_types:
    obj = cls.empty()
    pem = obj.pem
    assert isinstance(pem, str) and len(pem) > 0, f"{name}.pem failed"
    loaded = cls.loads(pem)
    assert isinstance(loaded, cls), f"{name}.loads() failed"

print(f"  ✓ All {len(pem_types)} types: pem + loads")

# ── 8. Password PEM (Encrypted) ──────────────────────────────

print("\n── 8. Password PEM (Encrypted) ──")

pw = b"test_password"
salt = b"test_salt_value"

x_pem_signer = root_a.x_pem(pw, salt)
loaded_signer = DilithiumSigner.x_loads(x_pem_signer, pw, salt)
assert loaded_signer.address == root_a.address, "password PEM round-trip address mismatch"
print("  ✓ DilithiumSigner x_pem + x_loads")

x_pem_kem = kem_a.x_pem(pw, salt)
loaded_kem = KyberFullKEM.x_loads(x_pem_kem, pw, salt)
assert loaded_kem.kyb_pubkey == kem_a.kyb_pubkey, "KEM password PEM round-trip mismatch"
print("  ✓ KyberFullKEM x_pem + x_loads")

# ── 9. Password Lock/Unlock ──────────────────────────────────

print("\n── 9. Password Lock/Unlock ──")

locked_signer = root_a.lock_with_password(pw, salt)
assert isinstance(locked_signer, XDilithiumSigner)
unlocked_signer = DilithiumSigner.unlock_with_password(locked_signer, pw, salt)
assert unlocked_signer.address == root_a.address
print("  ✓ DilithiumSigner lock + unlock")

locked_kem = kem_a.lock_with_password(pw, salt)
assert isinstance(locked_kem, XKyberFullKEM)
unlocked_kem = KyberFullKEM.unlock_with_password(locked_kem, pw, salt)
assert unlocked_kem.kyb_pubkey == kem_a.kyb_pubkey
print("  ✓ KyberFullKEM lock + unlock")

# ── 10. Session Handshake ─────────────────────────────────────

print("\n── 10. Session Handshake ──")

sb_a = SessionBuilder.new(root_b.address, delegate_a)
sb_b = SessionBuilder.new(root_a.address, delegate_b)
print(f"  SessionBuilder created for both parties")

session_a, session_b = perform_handshake(sb_a, sb_b)
print(f"  Handshake completed")

# ── 11. Encrypt / Decrypt ────────────────────────────────────

print("\n── 11. Encrypt / Decrypt ──")

messages = [
    b"hello from A",
    b"",
    b"x" * 10000,
    bytes(range(256)),
]

for i, msg in enumerate(messages):
    ct = session_a.encrypt(msg)
    pt = session_b.decrypt(ct)
    assert pt == msg, f"message {i} decrypt mismatch"
    
    ct2 = session_b.encrypt(msg)
    pt2 = session_a.decrypt(ct2)
    assert pt2 == msg, f"message {i} reverse decrypt mismatch"

print(f"  ✓ {len(messages)} messages encrypted/decrypted both directions")

# ── 12. Session Lock/Unlock ───────────────────────────────────

print("\n── 12. Session Lock/Unlock ──")

locked_session = session_a.lock_with_password(pw, salt)
assert isinstance(locked_session, XAloecryptSession)
unlocked_session = AloecryptSession.unlock_with_password(locked_session, pw, salt)
ct = unlocked_session.encrypt(b"after unlock")
pt = session_b.decrypt(ct)
assert pt == b"after unlock"
print("  ✓ Session lock + unlock + encrypt/decrypt")

# ── 13. Timestamp ─────────────────────────────────────────────

print("\n── 13. Timestamp ──")

ts = _ts_bytes_now()
ts_ms = int.from_bytes(ts, 'little')
dt = datetime.fromtimestamp(ts_ms / 1000)
print(f"  Current time: {dt}")
assert dt.year >= 2025
print("  ✓ Timestamp valid")

# ── Summary ───────────────────────────────────────────────────

print("\n" + "=" * 60)
print("ALL TESTS PASSED")
print("=" * 60)
```
